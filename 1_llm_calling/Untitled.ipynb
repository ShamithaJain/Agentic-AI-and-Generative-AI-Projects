```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

```python
llm= ChatGoogleGenerativeAI(model="gemini-2.5-flash")

#llm calling
response=llm.invoke("How many moons does earth have?")
print(response.text)
```

```python
response=llm.invoke([
    "system","You are a helpful assistant that answers in short one line",
    "human","How many moons does earth have"
])
print(response.text)
```

```python
prompt="Single tagline for a coffee shop only output the tagline"
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",temperature=0)
response=llm.invoke(prompt)
print(response)
```

## Call LLM Using Groq Models

```python
from langchain_groq import ChatGroq
```

```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq

load_dotenv()

llm = ChatGroq(
    model="qwen/qwen3-32b",
    reasoning_format="parsed"
)

messages = [
    ("system", "You are a helpful assistant that gives a short answer."),
    ("human", "How many moons does Jupiter have?")
]

response = llm.invoke(messages)

print(response.content)
```
